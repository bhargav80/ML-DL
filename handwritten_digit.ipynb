```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
(train_data,train_labels),(test_data,test_labels) = mnist.load_data()
```

```python
train_data[0].shape
```

```python
print(f"Training sample:\n{train_data[0]}\n")
print(f"Training sample:\n{train_labels[0]}\n")
```

```python
#Plot a single sample
plt.imshow(train_data[2]),train_labels[2]
```

```python
class_names = ['zero', 'one', 'two', 'three', 'four',
               'five', 'Six', 'Seven', 'eight', 'nine']
```

## Building the Model

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])

model_1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(),
                metrics=["accuracy"])
history_1 = model_1.fit(train_data,train_labels,
                        epochs=10,
                          validation_data=(test_data, test_labels))
```

##Normalizing

```python
train_data_norm = train_data/255.
test_data_norm = test_data/255.
```

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])

model_2.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001),
                metrics=["accuracy"])
history_2= model_2.fit(train_data_norm,train_labels,
                        epochs=10,
                          validation_data=(test_data_norm, test_labels))
```

```python
import pandas as pd
pd.DataFrame(history_1.history).plot(title="Non-normalized data")
pd.DataFrame(history_2.history).plot(title="Normalized data")
```

## Evaluating

```python
# Make some predictions with our models
y_probs = model_2.predict(test_data_norm)
#View the first 5 predictions
y_probs[0]
```

```python
class_names[tf.argmax(y_probs[10])]
```

```python
import random
def plot_random_image(model,images,true_labels,classes):
  i = random.randint(0,len(images))

  #Create predictions and targets
  target_image = images[i]
  pred_probs = model.predict(target_image.reshape(1, 28, 28))
  pred_label = classes[pred_probs.argmax()]
  true_label = classes[true_labels[i]]

  # Plot the target image
  plt.imshow(target_image, cmap=plt.cm.binary)

  # Change the color of the titles depending on if the prediction is right or wrong
  if pred_label == true_label:
    color = "green"
  else:
    color = "red"

  # Add xlabel information (prediction/true label)
  plt.xlabel("Pred: {} {:2.0f}% (True: {})".format(pred_label,
                                                   100*tf.reduce_max(pred_probs),
                                                   true_label),
             color=color) # set the color to green or red



```

```python
 #Check out a random image as well as its prediction
plot_random_image(model=model_2,
                  images=test_data_norm,
                  true_labels=test_labels,
                  classes=class_names)
```
