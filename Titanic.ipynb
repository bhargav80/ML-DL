```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('train.csv')
dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch']
print(dataset.head())


```

```python
feature_columns = [2,4,5,9,12]
dp = ['Survived']
X = dataset.iloc[:,feature_columns].values
y = dataset.iloc[:,1].values

print(X)
```

```python
print(y)
```

```python
missing_values = dataset.isnull().sum()
print(missing_values)
```

Handling Missing Values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,2:3])
X[:,2:3] = imputer.transform(X[:,2:3])
```

```python
print(X)
```

Encoding


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
X[:,1] = label_encoder.fit_transform(X[:,1])
# Print the first few rows of X and y after label encoding
print("X (after label encoding):")
print(X[:5])  # Print the first 5 rows of X
print("y:")
print(y[:5])  # Print the first 5 rows of y
len(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=1)

```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p =2)  #default neighors 5
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```
